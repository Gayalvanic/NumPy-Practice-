# --------------------------------------------------------------

##### 1. Inserting
##### Syntax: np.insert(array_name, index_pos, value, axis=Noneor0or1)

```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])
print(arr)
arr_1 = np.insert(arr, 2, 10)
print(arr_1)
```

```python
arr_2d = np.array([[1,2],
                   [3,4]])
print(arr_2d)
arr_2d_1 = np.insert(arr_2d, 1, [5,6], axis=1)  # inserting row
print(arr_2d_1)
```

#### 2. appending
##### Syntax :- np.append(array_name, value)

```python
arr = np.array([1, 2, 3, 4, 5, 6])
print(arr)
arr_1 = np.append(arr, [7, 8, 9])
print(arr_1)
```

```python
arr_2d = np.array([[1,2],
                   [3,4]])
print(arr_2d)
arr_2d_1 = np.append(arr_2d,[5,6])  # inserting row
print(arr_2d_1)
```

#### 3. Concatenate
##### Syntax: np.concatenate((array1, array2), axis = 0 or 1)

```python
arr = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
arr_1 = np.concatenate((arr2, arr), axis = 0)
print(arr_1)
```

```python
a = np.array([[1, 2],
              [3, 4]])

b = np.array([[5, 6],
              [7, 8]])

# Concatenate along columns
result = np.concatenate((a, b), axis=0)
print(result)
```

```python
a = np.array([[1, 2],
              [3, 4]])

b = np.array([[5, 6],
              [7, 8]])

# Concatenate along columns
result = np.concatenate((a, b), axis=1)
print(result)
```

#### 4. Deleting
##### Syntax: np.delete(array_name, index)

```python
arr = np.array([1, 2, 3])
arr_1 = np.delete(arr, 2)
print(arr_1)
```

```python
b = np.array([[5, 6],
              [7, 8]])

# Delete first column
result = np.delete(b, 0, axis=1)
print(result)
```

#### 5. Stacking and Splitting
##### we have np.vstack(), np.hstack() for stacking and np.vsplit(), np.hsplit()
##### Syntax: For Stacking - .vstack(array1, array2) "vertical stacking" .hstack(array1, array2)"horizontal stacking"
#####         For Splitting - .vsplit(array_name, no of times to split) "vertical split" .hsplit()"horizontal split"

```python
a = np.array([[1, 2],
              [3, 4]])
b = np.array([[5, 6],
              [7, 8]])
result = np.hstack((a, b))
print(result)
result = np.vstack((a, b))
print(result)
```

```python
a = np.array([1, 2, 3, 4, 5, 6])
result_h = np.hsplit(a, 3)
print(result_h)
a = np.array([1, 2, 3, 4, 5, 6]).reshape(3, 2)
result_v = np.vsplit(a, 3)
print(result_v)
```
