## Properties and Operations
### Shape:  shape returns the dimensions of a NumPy array as a tuple, showing how many elements it has along each axis.
### Size:  it returns the number of elements.
### ndim:  ndim stands for no of dimentions. this attribute returns the number of dimentions of an array. 
### dtype:  data type is returned.
### the above attributes are used to check certain things about the array.
### astype:  this is an operation used to convert one datatype to another


##### Shape 

```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])
print(arr.shape)

arr2 = np.array([[1, 2, 3],
                [4, 5, 6]])
print(arr2.shape)

arr3 = np.array([[[1, 2],[3, 4]], 
                [[5, 6],[7, 8]]])
                # the above list has 2 blocks, each block har 2 rows and two colomns hence 2x2x2
print(arr3)
print(arr3.shape)
```

#### Size


```python
print(arr.size)
print(arr2.size)
print(arr3.size)    
```

#### ndim

```python
print(arr.ndim)
print(arr2.ndim)
print(arr3.ndim)
```

#### dtype

```python
arr4 = np.array([1, 2, 3, 4])
print(arr4.dtype)
arr5 = np.array(['apple', 'banana', 'cherry'])
print(arr5.dtype)
arr6 = np.array([1.5, 2.5, 3.5])
print(arr6.dtype)
```

#### astype
##### Syntax: array.astype(new_type)

```python
arr_inttype = arr6.astype(int)
print(arr_inttype)
print(arr_inttype.dtype)
```

### operations on numpy arrays can be easily done without using loops.
#### Arithmetic Operations: + , - , * , ** , / , // , %

```python
math_arr = np.array([1,2,3,4,5])
print(math_arr + 2)
print(math_arr * 2)
print(math_arr / 2)
print(math_arr % 2)
```

#### Aggregate functions
##### sum, mean, min, max, standard deviation, varience.

```python
arr = np.array([1,2,3,4,5])

print("the sum of elements: ", np.sum(arr))
print("the mean of elements: ", np.mean(arr))
print("the smallest of elements: ", np.min(arr))
print("the largest element: ", np.max(arr))
print("Standard deviation is: ", np.std(arr))
print("Varience is: ",np.var(arr))
```
